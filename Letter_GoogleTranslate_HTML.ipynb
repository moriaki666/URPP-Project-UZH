```python
df_2['Brief ID'] = df_2['Title'].str.extract((r'Brief\s(\d+)'))
```

```python
df_2 = df_2.drop('File', axis=1)
```

```python
df_2.info()
df_2.head()
```

```python
import pandas as pd
from googletrans import Translator

# Assuming your dataframe is called 'df'

# Initialize the translator
translator = Translator()

# Function to translate text
def translate_text(text, dest_language='en'):
    try:
        return translator.translate(text, dest=dest_language).text
    except:
        return text  # Return original text if translation fails

# Apply translation to the 'transcription' column
df_2['translated_transcription'] = df_2['Transcription'].apply(lambda x: translate_text(x, dest_language='en'))
```

```python
df_2 = df_2.sort_values(by=['Brief ID'], ascending=True)
```

```python
output_path = 'extracted_translated_letters.csv'  # Specify the output file path
df_2.to_csv(output_path, index=False)
```
